```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = pd.read_csv("hotel_bookings.csv")
```

## Heatmap

```python
sns.heatmap(data.isna(),annot=False,cbar=False,cmap="viridis")
data.isna().sum()
```

```python
data = data.drop("company",axis=1)
```

## For Children Column

```python
def children(x):
    if pd.isna(x) == True:
        return data.children.mean()
    else:
        return x


data["children"] = data["children"].apply(children)
```

## For Country Column

```python
def country(x):
    if pd.isna(x) == True:
        return "Unknown"
    else:
        return x
data["country"] = data["country"].apply(country)
```

## For Agent Column

```python
def agent(x):
    if pd.isna(x) == True:
        return data.agent.mean()
    else:
        return x


data["agent"] = data["agent"].apply(agent)
```
